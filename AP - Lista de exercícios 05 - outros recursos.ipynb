### **01) Execute as seguintes tarefas**:

1.   Crie uma função `primo` para calcular se um número passado como parâmetro é primo. A função deve retornar `True` ou `False`
2.   Crie uma função `divisores` para retornar os divisores de um número passado como parâmetro. A função deve retornar a lista dos divisores.
3.   Crie uma função `mdc` para calcular o máximo divisor comum (MDC) de uma lista de números passada como parâmetro 
4.   Crie uma função `mmc` para calcular o mínimo múltiplo comum (MMC) de uma lista de números passada como parâmetro 

**Saída esperada**:
```
primo(1): False
primo(7): True
primo(52): False
primo(53): True

divisores(6): [1, 2, 3, 6]
divisores(24): [1, 2, 3, 4, 6, 8, 12, 24]
divisores(52): [1, 2, 4, 13, 26, 52]
divisores(53): [1, 53]

MDC(16, 24, 32): 8
MDC(30, 90, 180): 30
MDC(30, 36, 72): 6
MDC(9, 45, 180): 9

MMC(8, 12, 28): 168
MMC(30, 36, 40): 360
MMC(4, 5, 9): 180
MMC(3, 5, 6): 30
```



````python
# exercício 01 - 1
def numberIsPrime(n):
    if (n <= 1):
      return False
    if (n <= 3):
      return True

    if (n % 2 == 0 or n % 3 == 0):
      return False
  
    i = 5
    while(i * i <= n) : 
      if (n % i == 0 or n % (i + 2) == 0):
        return False
      i = i + 6
    return True

def printNumberIsPrime(n):
  print("primo(" + str(n) + "): " + str(numberIsPrime(n)))

printNumberIsPrime(1)
printNumberIsPrime(7)
printNumberIsPrime(52)
printNumberIsPrime(53)

print()



# exercício 01 - 2

def divisors(n):
    divisors = list()
    i = 1
    while i <= n:
      if (n % i == 0):
        divisors.append(i)
      i = i + 1
    return divisors

def printDivisors(n):
  print("divisores(" + str(n) + "): " + str(divisors(n)))

printDivisors(6)
printDivisors(24)
printDivisors(52)
printDivisors(53)

print()



# exercício 01 - 3

def gcd(nTuple):
  result = nTuple[0]
  for x in nTuple[1:]:
    if (result < x):
      temp = result
      result = x
      x = temp
    while (x != 0):
      temp = x
      x = result % x
      result = temp
  return result

def printGcd(*nTuple):
  print("MDC" + str(nTuple) + ": " + str(gcd(nTuple)))

printGcd(16, 24, 32)
printGcd(30, 90, 180)
printGcd(30, 36, 72)
printGcd(9, 45, 180)

print()



# exercício 01 - 4

def lcm(nTuple):
  nList = list(nTuple)
  nList.sort(reverse = True)
  x = nList[0]

  while (True):
    sum = 0
    for i in nList:
      if x % i == 0:
        sum += 1
    if sum == len(nList):
      break
    else:
      x += 1

  return x

def printLcm(*nTuple):
  print("MMC" + str(nTuple) + ": " + str(lcm(nTuple)))

printLcm(8, 12, 28)
printLcm(30, 36, 40)
printLcm(4, 5, 9)
printLcm(3, 5, 6)
````

2. A função `busca(chave, lista)` deve executar uma busca binária, utilizando uma outra função recursiva chamada `busca_binaria(ini, fim)`. 
3. A busca binária divide o espaço de busca pela metade a cada passo de execução. A ideia geral é comparar a chave com a posição média entre os valores `ini` e `fim`. Caso o valor da chave seja maior do a chave da posição média (meio), faz-se uma nova chamada com os índices `(meio+1, fim)` e caso contrário, com os índices `(ini, meio)`. 
Veja os exemplos das chamadas recursivas abaixo:
```

busca_binaria(0,10)
busca_binaria(6,10)
busca_binaria(6,8)

```

4. Para entender o funcionamento da busca binária, vamos usar uma brincadeira como exemplo. Peça outra pessoa para escolher um número de 1 a 100. Você vai advinhar esse número aplicando os conceitos da busca binária, perguntando da seguinte forma:
```
Se o número escolhido fosse o número 81, as chamadas recursivas seriam as seguintes 
(supondo que o número 1 estaria na posição 0 da lista e o número 100 na posição 99):
busca_binaria(0,99)
busca_binaria(51,99)
busca_binaria(76,99)
busca_binaria(76,88)
busca_binaria(76,82)
busca_binaria(80,82)
busca_binaria(80,81)
busca_binaria(80,80)
```
**Saída esperada**:
```
abel: 0
pedro: 6
renata: 7
ze: 10
nao_estou_na_lista: -1
```



````python
# exercício 02

def busca(chave, lista):
  result = busca_binaria(lista, 0, len(lista) - 1, chave)

  if result != -1: 
    print(chave + ": " + str(result))
  else: 
    print("nao_estou_na_lista: " + str(result))

def busca_binaria(lista, esq, dir, chave):
  if dir >= esq:
    mei = esq + (dir - esq) // 2

    if (lista[mei] == chave):
      return mei 

    elif (lista[mei] > chave):
      return busca_binaria(lista, esq, mei - 1, chave)
    else:
      return busca_binaria(lista, mei + 1, dir, chave)
  else:
    return -1


lista = ['abel', 'clara', 'joao', 'lucas', 'luciana', 'maria', 'pedro', 'renata', 'renato', 'rosana', 'ze']

busca("abel", lista)
busca("pedro", lista)
busca("renata", lista)
busca("ze", lista)
busca("undefined", lista)
````

6. A saída do algoritmo deve ser o conteúdo do arquivo `saida.txt`

````python
# exercício 03

# definir esturutras de dados que precisar 

qIds = dict()
qClasses = dict()
qFeatures = dict()
dados = list()
saida = list()

with open('/content/drive/My Drive/dados.txt') as arquivo: 
  for linha in arquivo:
    dados.append(linha.split(' '))

# Lógica para montagem do arquivo de saída
for linha in dados:
  # Dicionário com chave composta: id e classe
  if (linha[1] not in qClasses):
    qClasses[linha[1]] = {0: 0, 1: 0, 2: 0}

  classInt = int(linha[0], 10)
  if (classInt not in qClasses[linha[1]]):
    qClasses[linha[1]][classInt] = 1
  else:
    qClasses[linha[1]][classInt] = qClasses[linha[1]][classInt] + 1

  # Dicionário com chave simples: id
  if (linha[1] not in qIds):
    qIds[linha[1]] = 1
  else:
    qIds[linha[1]] = qIds[linha[1]] + 1

  # Dicionário com chave composta: id e numero da feature
  if (linha[1] not in qFeatures):
    qFeatures[linha[1]] = {5: 0, 10: 0, 15: 0, 20: 0, 25: 0, 30: 0, 35: 0, 40: 0, 45: 0}

  for i in range(len(linha) - 3):
    splittedFeature = linha[i + 2].split(':')
    numeroFeature = int(splittedFeature[0], 10)
    valorFeature = float(splittedFeature[1])

    if (numeroFeature % 5 == 0 and numeroFeature != 0):
      qFeatures[linha[1]][numeroFeature] = qFeatures[linha[1]][numeroFeature] + valorFeature

# Arquivo de saída
for qId in qIds:
  for numeroFeature in qFeatures[qId]:
    valorFeature = qFeatures[qId][numeroFeature] / qIds[qId]
    valorFeature = float("{0:.4}".format(valorFeature))
    qFeatures[qId][numeroFeature] = valorFeature

  saida.append(qId.replace('qid', 'Query') + ' Total:' + str(qIds[qId]) + ' ' + str(qClasses.get(qId)).replace('{', '').replace('}', '').replace(' ', '').replace(',', ' ') + ' Features filtradas:' + str(list(qFeatures.get(qId).values())).replace(',', '').replace('[', '').replace(']', ''))

arquivo = open('/content/drive/My Drive/saida.txt', 'w')
for linha in saida:
  print(linha)
  arquivo.write(linha + '\n')
arquivo.close()

print('\nArquivo da saída atualizado.')
````

2. Criar uma classe herdada desta primeira, ilustrando os conceitos de herança, encapsulamento e polimorfismo

````python
# exercício 04 - <escrever o código neste bloco>
class Pessoa:
  def __init__(self, nome, idade):
    self.nome = nome
    self.idade = idade

  def bemVindo(self):
    pass


class Gerente(Pessoa):
  def __init__(self, nome, idade):
    super().__init__(nome, idade)

  def setSalario(self, salario):
    self.__salario = salario

  def getSalario(self, salario):
    return self.__salario

  def bemVindo(self):
    print("BEM VINDO\nNome do Gerente: " + self.nome + "\nIdade do Gerente: " + str(self.idade) + "\nSalário do Gerente: " + str(self.__salario))


class Vendedor(Pessoa):
  def __init__(self, nome, idade):
    super().__init__(nome, idade)

  def setSalario(self, salario):
    self.__salario = salario

  def getSalario(self, salario):
    return self.__salario

  def bemVindo(self):
    print("BEM VINDO\nNome do Vendedor: " + self.nome + "\nIdade do Vendedor: " + str(self.idade) + "\nSalário do Vendedor: " + str(self.__salario))


def exibirPessoa(pessoa):
  pessoa.bemVindo()


gerente = Gerente("Gerente1", 30)
gerente.setSalario(8000.0)
exibirPessoa(gerente)

print()

vendedor = Vendedor("Vendedor1", 20)
vendedor.setSalario(4000.0)
exibirPessoa(vendedor)
````
