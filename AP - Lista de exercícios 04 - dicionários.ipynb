### **01) Execute as seguintes tarefas**:

1.   Crie um dicionário vazio para armazenar o preço dos produtos encontrados na feira
2.   Insira os produtos abaixo no dicionário (produto, preço)
```
banana, 6.50
maçã, 8.0 
uva 12.0
``` 
3.   Imprima cada produto e o seu preço
4.   Aumente o preço de cada produto em 10% (mantenha o valor anterior usando um único dicionário!)
5.   Imprima novamente cada produto, o preço anterior e o preço alterado 

**Saída esperada**:
```
banana: 6.50
maçã: 8.00
uva: 12.00

banana: preço anterior 6.50 preço atual 7.15
maçã: preço anterior 8.00 preço atual 8.80
uva: preço anterior 12.00 preço atual 13.20
```



````python
# exercício 01
frutas = dict([('banana', [6.50]), ('maçã', [8.0]), ('uva', [12.0])])

for produto, preco in frutas.items():
    print(produto + ':', format(preco[0], '.2f'))
print()

for produto, preco in frutas.items():
    preco.append(float(format(preco[0] * 1.1, '.2f')))

for produto, preco in frutas.items():
    print(produto + ': preço anterior ' + format(preco[0], '.2f') + ' preço atual', format(preco[1], '.2f'))
````

### **02) Execute as seguintes tarefas**:

1.   Crie um dicionário para armazenar a sua agenda de contatos
2.   Cada contato poderá ter gravados telefones dos tipos ***celular***, ***comercial*** ou ***residencial***
3.   Deve ser possível consultar todos os telefones de um contato, bem como consultar um telefone específico de um contato
*   Imprimir o telefone comercial do joao
*   Imprimir o telefone celular da maria
*   Imprimir todos os telefones do ze

4.   O seu código deve exibir a saída abaixo

**Saída esperada**:
```
joao: (31)1111-1111
maria: (31)9999-9992
telefones do ze: ['(31)9999-9993', '(31)2222-1111']

Nome:   joao
        Telefone celular:	    (31)9999-9991
	    Telefone comercial:      (31)1111-1111

Nome:   maria
	    Telefone celular:        (31)9999-9992

Nome:   ze
	    Telefone celular:	    (31)9999-9993
	    Telefone residencial:	(31)2222-1111

```



````python
# exercício 02
contatos = dict([
    (
        'joao',
        {
            'celular': '(31)9999-9991',
            'comercial': '(31)1111-1111',
            'residencial': None
        }
    ),
    (
        'maria',
        {
            'celular': '(31)9999-9992',
            'comercial': None,
            'residencial': None
        }
    ),
    (
        'ze',
        {
            'celular': '(31)9999-9993',
            'comercial': None,
            'residencial': '(31)2222-1111'
        }
    )
])

nome = 'joao'
print(nome + ':', contatos.get(nome).get('comercial'))

nome = 'maria'
print(nome + ':', contatos.get(nome).get('celular'))

nome = 'ze'
print(nome + ':', list(filter(None, contatos.get(nome).values())))

print()
for nome, telefone in contatos.items():
    print('Nome: \t' + nome)

    tel = telefone.get('celular')
    if (tel is not None):
        print('\tTelefone celular: ' + tel)

    tel = telefone.get('comercial')
    if (tel is not None):
        print('\tTelefone comercial: ' + tel)

    tel = telefone.get('residencial')
    if (tel is not None):
        print('\tTelefone residencial: ' + tel)

    print()
````

### **03) Execute as seguintes tarefas**:

1.   Dado o dicionário abaixo:
```
docs = { 
'documento 1' : 'palavras em python sobre dicionário em python', 
'documento 2' : 'palavras de dicionário de dicionário significa palavras de língua', 
'documento 3' : 'língua em dicionário significa sobre língua'
}
```
2.   Crie as estruturas de dados que julgar necessário para exibir a saída abaixo

**Saída esperada**:
```
LISTAR AS PALAVRAS POR DOCUMENTO:

- documento 1 (total = 7 palavra(s)): 
   palavras aparece 1 vez(es)
   em aparece 2 vez(es)
   python aparece 2 vez(es)
   sobre aparece 1 vez(es)
   dicionário aparece 1 vez(es)

- documento 2 (total = 9 palavra(s)): 
   palavras aparece 2 vez(es)
   de aparece 3 vez(es)
   dicionário aparece 2 vez(es)
   significa aparece 1 vez(es)
   língua aparece 1 vez(es)

- documento 3 (total = 6 palavra(s)): 
   língua aparece 2 vez(es)
   em aparece 1 vez(es)
   dicionário aparece 1 vez(es)
   significa aparece 1 vez(es)
   sobre aparece 1 vez(es)

LISTAR ONDE AS PALAVRAS APARECEM:

- palavras (total = 3 ocorrência(s) e 2 documento(s)): 
   aparece 1 vez(es) no documento 1
   aparece 2 vez(es) no documento 2

- em (total = 3 ocorrência(s) e 2 documento(s)): 
   aparece 2 vez(es) no documento 1
   aparece 1 vez(es) no documento 3

- python (total = 2 ocorrência(s) e 1 documento(s)): 
   aparece 2 vez(es) no documento 1

- sobre (total = 2 ocorrência(s) e 2 documento(s)): 
   aparece 1 vez(es) no documento 1
   aparece 1 vez(es) no documento 3

- dicionário (total = 4 ocorrência(s) e 3 documento(s)): 
   aparece 1 vez(es) no documento 1
   aparece 2 vez(es) no documento 2
   aparece 1 vez(es) no documento 3

- de (total = 3 ocorrência(s) e 1 documento(s)): 
   aparece 3 vez(es) no documento 2

- significa (total = 2 ocorrência(s) e 2 documento(s)): 
   aparece 1 vez(es) no documento 2
   aparece 1 vez(es) no documento 3

- língua (total = 3 ocorrência(s) e 2 documento(s)): 
   aparece 1 vez(es) no documento 2
   aparece 2 vez(es) no documento 3
```




````python
# exercício 03

docs = { 
    'documento 1': 'palavras em python sobre dicionário em python', 
    'documento 2': 'palavras de dicionário de dicionário significa palavras de língua', 
    'documento 3': 'língua em dicionário significa sobre língua'
}


print('LISTAR AS PALAVRAS POR DOCUMENTO: \n')
setTodasPalavras = set()

for documento, palavras in docs.items():
    arrayPalavras = palavras.split(' ')
    print('- ' + documento + ' (total = ' + str(len(arrayPalavras)) + ' palavra(s)): ')

    arrayPalavrasSemDuplicatas = list(dict.fromkeys(arrayPalavras))
    for i in range(len(arrayPalavrasSemDuplicatas)): 
        print(arrayPalavrasSemDuplicatas[i] + ' aparece ' + str(arrayPalavras.count(arrayPalavrasSemDuplicatas[i])) + ' vez(es)') 
        setTodasPalavras.add(arrayPalavrasSemDuplicatas[i])
    print()


print('LISTAR ONDE AS PALAVRAS APARECEM: \n')
arrayTodasPalavras = list(setTodasPalavras)

for i in range(len(arrayTodasPalavras)):
    totalOcorrencias = 0
    contadorDocumentos = 0

    for documento, palavras in docs.items():
        totalOcorrencias += palavras.count(arrayTodasPalavras[i])
        possuiPalavras = palavras.count(arrayTodasPalavras[i])
        if (possuiPalavras != 0):
            contadorDocumentos += 1
    
    print('- ' + arrayTodasPalavras[i] + ' (total = ' + str(totalOcorrencias) + ' ocorrência(s) e ' + str(contadorDocumentos) + ' documento(s)): ')

    for documento, palavras in docs.items():
        ocorrencias = palavras.count(arrayTodasPalavras[i])
        if (ocorrencias != 0):
            print('aparece ' + str(ocorrencias) + ' vez(es) no ' + documento)
    print()
````
